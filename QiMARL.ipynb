```python
import numpy as np
import pandas as pd
import time
import matplotlib.pyplot as plt
#from qiskit import QuantumCircuit, Aer, transpile, assemble
from qiskit_algorithms.optimizers import COBYLA
from scipy.optimize import minimize
from qiskit_aer import Aer
from qiskit.visualization import circuit_drawer
```

```python
# Generate synthetic energy data
np.random.seed(42)
hours = 20
power_stations = ["Gen_Station_1", "Gen_Station_2", "Gen_Station_3"]
nodal_stations = ["Nodal_Station_1", "Nodal_Station_2", "Nodal_Station_3"]

data = {
    "Timestamp": pd.date_range(start="2025-02-01", periods=hours, freq="H")
}
for ps in power_stations:
    data[f"{ps} (MW)"] = np.random.normal(loc=100 + 10 * power_stations.index(ps), scale=10, size=hours)
for ns in nodal_stations:
    data[f"{ns} (MW)"] = np.random.normal(loc=80 + 10 * nodal_stations.index(ns), scale=8, size=hours)

historical_data = pd.DataFrame(data)
historical_data.to_csv("synthetic_energy_data.csv", index=False)

# Load historical data
train_data = historical_data[:15]
test_data = historical_data[15:]

# Quantum rotation angles from prior research
def update_angles(epoch):
    return 1.57 - 0.01 * epoch, 0.785 + 0.01 * epoch, 1.25 - 0.005 * epoch

# Quantum circuit initialization for 9 qubits
def initialize_quantum_state(entangle_pattern):
    qc = QuantumCircuit(9)
    qc.h(range(9))  # Superposition across all 9 qubits
    for i, j in entangle_pattern:
        qc.cx(i, j)  # Introduce dynamic entanglement pattern
    return qc

# Encode supply and demand data into quantum states
def encode_data(qc, supply, demand):
    for i, (s, d) in enumerate(zip(supply, demand)):
        angle = np.arctan(d / max(s, 1))  # Prevent division by zero
        qc.ry(angle, i)

# Apply optimized quantum rotation gates
def apply_rotation_gates(qc, theta, phi, lambda_):
    for qubit in range(9):
        qc.rx(theta, qubit)
        qc.ry(phi, qubit)
        qc.rz(lambda_, qubit)

# Measure quantum state
def measure_quantum_state(qc):
    qc.measure_all()
    simulator = Aer.get_backend("aer_simulator")
    transpiled_qc = transpile(qc, simulator)
    results = simulator.run(transpiled_qc).result()
    return results.get_counts()

# Local reward function (per row allocation)
def local_reward(energy_alloc, demand, prev_deficit):
    deficit = demand - energy_alloc
    return -np.abs(deficit) if deficit > prev_deficit else 1 / (1 + np.abs(deficit))

# Global reward function (per epoch)
def global_reward(convergence_time, prev_convergence_times, prev_rewards):
    avg_prev_time = np.mean(prev_convergence_times) if prev_convergence_times else convergence_time
    avg_prev_reward = np.mean(prev_rewards) if prev_rewards else 0
    time_penalty = 1 / (1 + convergence_time / avg_prev_time)  # Higher time, lower reward
    return avg_prev_reward * time_penalty

# Optimize policy using reinforcement learning
def optimize_policy(train_data, epochs=10):
    best_policy = {}
    angle_history = []
    reward_history = []
    local_rewards = []
    global_rewards = []
    convergence_times = []
    entangle_patterns = [
        [(0, 1), (3, 4), (6, 7)],
        [(1, 2), (4, 5), (7, 8)],
        [(0, 2), (3, 5), (6, 8)],
    ]
    prev_convergence_times = []
    prev_rewards = []
    for epoch in range(epochs):
        start_time = time.time()
        theta, phi, lambda_ = update_angles(epoch)
        angle_history.append((epoch, theta, phi, lambda_))
        total_epoch_reward = 0
        prev_deficit = float("inf")
        for _, row in train_data.iterrows():
            supply = [row[f"{ps} (MW)"] for ps in power_stations]
            demand = [row[f"{ns} (MW)"] for ns in nodal_stations]
            qc = initialize_quantum_state(entangle_patterns[epoch % len(entangle_patterns)])
            encode_data(qc, supply, demand)
            apply_rotation_gates(qc, theta, phi, lambda_)
            result = measure_quantum_state(qc)
            optimized_energy = sum(supply) * max(result.values()) / sum(result.values())
            local_rwd = local_reward(optimized_energy, sum(demand), prev_deficit)
            total_epoch_reward += local_rwd
            local_rewards.append(local_rwd)
            prev_deficit = sum(demand) - optimized_energy
        convergence_time = time.time() - start_time
        global_rwd = global_reward(convergence_time, prev_convergence_times, prev_rewards)
        global_rewards.append(global_rwd)
        total_epoch_reward += global_rwd
        reward_history.append(total_epoch_reward)
        prev_convergence_times.append(convergence_time)
        prev_rewards.append(total_epoch_reward)
        convergence_times.append(convergence_time)
        print(f"Quantum Circuit for Epoch {epoch}:")
        print(qc.draw())
    return reward_history, local_rewards, global_rewards, convergence_times, angle_history

# Run optimization
reward_history, local_rewards, global_rewards, convergence_times, angle_history = optimize_policy(train_data)

# Plot all required metrics
plt.figure(figsize=(15, 10))
plt.subplot(2, 3, 1)
plt.plot(local_rewards, marker='o')
plt.xlabel("Data Points")
plt.ylabel("Local Reward")
plt.title("Local Reward vs Data Points")

plt.subplot(2, 3, 2)
plt.plot(global_rewards, marker='s')
plt.xlabel("Epoch")
plt.ylabel("Global Reward")
plt.title("Global Reward vs Epoch")

plt.subplot(2, 3, 3)
plt.plot(reward_history, marker='^')
plt.xlabel("Epoch")
plt.ylabel("Overall Reward")
plt.title("Overall Reward vs Epoch")

plt.subplot(2, 3, 4)
plt.plot(convergence_times, marker='*')
plt.xlabel("Epoch")
plt.ylabel("Convergence Time (s)")
plt.title("Convergence Time vs Epoch")

plt.subplot(2, 3, 5)
epochs, theta_vals, phi_vals, lambda_vals = zip(*angle_history)
plt.plot(epochs, theta_vals, label='Theta')
plt.plot(epochs, phi_vals, label='Phi')
plt.plot(epochs, lambda_vals, label='Lambda')
plt.xlabel("Epoch")
plt.ylabel("Rotation Angles")
plt.title("Angle Change vs Epoch")
plt.legend()
plt.show()

# Print best epochs
print(f"Lowest Convergence Time: {min(convergence_times)} seconds at Epoch {np.argmin(convergence_times)}")
print(f"Highest Overall Reward: {max(reward_history)} at Epoch {np.argmax(reward_history)}")
```
